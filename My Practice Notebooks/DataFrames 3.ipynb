### The .apply() Method with Row Values

90. Apply a Function to every DataFrame Row with the apply Method

```python
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
#how to combine values in a row in custom function
#what if we want to create a custom columns
#e.g we want to create a column of Ranking and rank movies

def good_movie(row):
    actor=row[1]
    budget=row[4]
    
    if actor == "Pierce Brosnan":
        return "The best"
    elif actor =="Roger Moore" and budget>40:
        return "Enjoyable"
    else:
        return "I have no clue"
    
#all row values passed as an array of values one by one   
#actor column taken and assigned to actor variable
#inside row square bracket, we provide index number of row column
#Actor value is at index position of 1

bond.apply(good_movie,axis="columns")

#this time we want to apply this method row wise
#since we are moving horizontally across all columns, we are moving left to right across 6 columns
#we are traversing through columns
#press shift and tab to see parameter
```

### The .copy() Method

91. Create a Copy of a DataFrame with the copy Method

```python
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
#creates an exact copy but stores in a different memory
#lets say we want to pull out Director column
bond["Director"]
```

```python
#assigning this series to a brand new variable
#this series represent a chunk of our dataset
directors=bond["Director"]
directors.head(3)
```

```python
#let's say we want to replace John Glen with Mister John Glen
directors["A View to a Kill"]
```

```python
directors["A View to a Kill"]="Mister John Glen"
#here we get an error, trying to write on copy of data
#even though value has been changed
```

```python
directors.head(3)
#value has been changed
```

```python
bond.head(3)
```

```python
#what if we dont want to overwrite the original values?
#in this case, we create a separate copy
```

```python
bond=pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
directors=bond["Director"].copy()
directors.head(3)
```

```python
directors["A View to a Kill"]="Mister John Glen"
```

```python
directors.head(3)
#this is a copy
```

```python
bond.head(3)
#value has not been overwritten
#writing done on sub-object,original dataset is preserved
```
