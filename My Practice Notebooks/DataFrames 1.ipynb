### Sort a DataFrame with the .sort_values() Method, Part I

```python
#sort_values method called upon a dataframe
nba.sort_values("Name")
#by parameter expects a string or string values
#press shift and tab to see
#by default it puts NaNs at last, as decided by na_position parameter
#entire dataframe is sorted using the Name column
```

```python
nba.sort_values("Name",ascending=False)
#ascending paramter
```

```python
nba.sort_values("Age")
#shortest age shown above
nba.sort_values("Age", ascending=False)
```

```python
nba.sort_values("Salary",ascending=False,inplace=True)
nba.head(3)
```

```python
#in case we dont fill NaN values, they will be placed at end
nba.sort_values("Salary").tail()
#na_position set to last
```

```python
nba.sort_values("Salary",ascending=False,na_position="first").head()
#at top we have Null salaries
```

```python
nba.sort_values("Salary",ascending=False,na_position="first").tail()
#smallest salary at last
```

### Rank Values with the .rank() Method

```python
nba=pd.read_csv("nba.csv")
nba.head(3)
```

```python
#here we get rid of all null values
nba=pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"]=nba["Salary"].fillna(0).astype("int")
nba.head(3)
```

```python
#Rank method is called on a single series
#let's say we want to rank the players with salaries
#highest salary gets a rank of 1
nba["Salary"].rank(ascending=False).astype("int")
```

```python
nba["Salary Rank"]=nba["Salary"].rank(ascending=False).astype("int")
nba.head(3)
```

```python
nba.sort_values(by="Salary",ascending=False)
#ranks sorted by salaries, highest shown first
```
