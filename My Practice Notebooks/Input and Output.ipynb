### Quick Object Conversions

```python
#how to convert pandas object such as series or daaframe to vanilla python object such as a list, dictionary or a string
```

```python
url="https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv"
baby_names=pd.read_csv(url)
baby_names.head()
```

```python
baby_names["Child's First Name"]
```

```python
#in order to convert a series to a dataframe, call to_frame method
baby_names["Child's First Name"].to_frame()
```

```python
baby_names["Child's First Name"].tolist()
# tolist method
```

```python
baby_names["Child's First Name"].to_dict()
#it takes index position and use it a key in dictionary
#pythons dict doesnt support duplicate values in keys
#make sure your index is unique before converting it into dictionary
```

```python
baby_names["Child's First Name"]
#lets say we want string with all the names and also have them sorted in alphabatical order
#we also want duplicates removed , and also want name properly capitalized
#we can pythons join method  and concatenate
"!".join(["a","b","c"])
```

```python
", ".join(baby_names["Child's First Name"])
```

```python
", ".join(baby_names["Child's First Name"].str.title())
```

```python
#check out working with text data section
#to remove duplicates
", ".join(baby_names["Child's First Name"].str.title().drop_duplicates())
```

```python
#in sorted order
", ".join(baby_names["Child's First Name"].str.title().drop_duplicates().sort_values())
```

### Export CSV File with the to_csv Method

```python
url="https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv"
baby_names=pd.read_csv(url)
baby_names.head()
```

```python
#we will look at how to export csv file using to csv method
baby_names.to_csv("NYC_Baby_Names.csv")
#you have to provide .csv in strings of file name
#avoid spaces in your filename
```

```python
#pandas writes data to a csv file
#go to folder to see the file
#click on it to view the file
#index of dataframe has been included
#index has no name assigned to it as seen by a comma
```

```python
baby_names.to_csv("NYC_Baby_Names.csv",index=False)
#set index to false to remove index values
#be careful with filenames and not write to any original filename
```

```python
#how to customize columns to include
baby_names.to_csv("NYC_Baby_Names.csv",index=False,columns=["Gender","Ethnicity","Child's First Name"])
```

```python
#if you run into a bug 
#add a parameter called encoding
#you can also look for other encoding format than utf-8 using google search
baby_names.to_csv("NYC_Baby_Names.csv",index=False,columns=["Gender","Ethnicity","Child's First Name"], encoding="utf-8")
```

### Export Excel File

157. Export Excel File with the to_excel Method

```python
url="https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv"
baby_names=pd.read_csv(url)
baby_names.head()
```

```python
#we will take pandas objects and how to write to excel
```

```python
#we will create two datasets
baby_names["Gender"]
```

```python
baby_names["Gender"]=="FEMALE"
```

```python
baby_names[baby_names["Gender"]=="FEMALE"]
```

```python
girls=baby_names[baby_names["Gender"]=="FEMALE"]
```

```python
boys=baby_names[baby_names["Gender"]=="MALE"]
```

```python
#now we have two dataframes
#process is going to be a little complex
#we need to create an excel writer object
#it is a container to export one or more worksheets
excel_file=pd.ExcelWriter("Baby_Names.xlsx")
#give extension name as well
```

```python
#take a dataframe and call toexcel method on it
girls.to_excel(excel_file,sheet_name="Girls",index=False)
```

```python
boys.to_excel(excel_file,sheet_name="Boys",index=False,columns=["Year of Birth","Gender","Ethnicity"])
```

```python
#we still havent got an excel file
#now we will call save method on excel writer object
excel_file.save()
```
