### The precision Option

```python
#alters number of digits after floating point numbers
#it doesnt modify the number, just hide complexity of it
df=pd.DataFrame(np.random.randn(5,5))
df
```

```python
#by default pandas show six digits or numbers after decimal point
#we want to limit to 2 in display only
#we dont want to modify actual numbers
```

```python
pd.get_option("precision")
```

```python
pd.set_option("precision",2)
```

```python
df
#values rounded off
```

```python
pd.reset_option("precision")
```

```python
df
#original data is preserved
```
