### The pd.concat Method, Part 1

124. The pd.concat Method, Part 1

```python
week1=pd.read_csv("Restaurant - Week 1 Sales.csv")
week2=pd.read_csv("Restaurant - Week 2 Sales.csv")
customers=pd.read_csv("Restaurant - Customers.csv")
foods=pd.read_csv("Restaurant - Foods.csv")
```

```python
#we will look at pd.concat method that combines two or more dataframes
```

```python
#week1 and week2 dataframes are quite similar, listing sales for that week
week1.head(3)
```

```python
week2.head(3)
```

```python
#what if we want to combine week1 and week2 
#both have same column names
#both dataframes have 250 rows, in theory after concatenation, we should get 500 rows
#we will also see another method to concatenate
```

```python
pd.concat(objs=[week1,week2])
#shift plus tab to see parameters
#objs short for objects, can be a list of dataframes
#order that you pass will be the way they appear
#we get 500 rows, but why index position end at 249?
#by default pandas take index of week2 and also glue it to index at end of week1
```

```python
pd.concat(objs=[week1,week2],ignore_index=True)
#there is an additional parameter called ignore index
#new index created from scratch
#depends if you want to keep original index or not
```

```python
#we can also call append method on each of our dataframe
week1.append(other=week2)
```

```python
week1.append(other=week2,ignore_index=True)
#this always create a new dataframe, no inplace parameter here
```

### Inner Joins, Part 1

126. Inner Joins, Part 1

```python
week1=pd.read_csv("Restaurant - Week 1 Sales.csv")
week2=pd.read_csv("Restaurant - Week 2 Sales.csv")
customers=pd.read_csv("Restaurant - Customers.csv")
foods=pd.read_csv("Restaurant - Foods.csv")
```

```python
#it merges two datasets based on commonly shared values
#the point where datasets overlap is inner join
#there are batch of customers in both week1 and week2 that will be included in inner join or an inner merge
```

```python
week1.head(3)
```

```python
week2.head(3)
```

```python
#we are lucky here that both weeks have same column name
#if we had different names in columns, we would get an issue 
#and would have to use different parameters, we will look at it later
```

```python
#we call merge method on either one
week1.merge(week2,how="inner",on="Customer ID")
#you can think of dataframes side by side, week1 is on left dataframe, and week2 is on right
#how parameter describes how
#on parameter only works if both columns have same name
#here we get new dataframe, Food ID_x represents week1 and Food ID_y is week2
#to make column IDs unique, it gives it _x and _y names
#we will change it later
#some customer id repeated
```

```python
week1.merge(week2,how="inner",on="Customer ID").head(3)
```

```python
week1["Customer ID"]==155
```

```python
week1[week1["Customer ID"]==155]
#we will do filteration to customer ID
#we have two occurances of customer ID in week 1
```

```python
week2[week2["Customer ID"]==155]
# in second week, this customer came once
#pandas repeats the process and created two outputs
#it doesnot de-duplicate two values
```

```python
week1.merge(week2,how="inner",on="Customer ID",suffixes=["- Week 1"," - Week 2"])
#to remove default column names with x and y
```

### Outer Joins

128. Outer Joins

```python
week1=pd.read_csv("Restaurant - Week 1 Sales.csv")
week2=pd.read_csv("Restaurant - Week 2 Sales.csv")
customers=pd.read_csv("Restaurant - Customers.csv")
foods=pd.read_csv("Restaurant - Foods.csv")
```

```python
#outer joins combines values in either dataframes or both
#another can be values in eith A or B but not both
```

```python
week1.head(2)
```

```python
week2.head(2)
```

```python
week1.merge(week2,how="outer",on="Customer ID")
```

```python
week1.merge(week2,how="outer",on="Customer ID",suffixes=["- Week1", " - Week2"])
#customer 537 came and ordered food 9 on week1 and food 5 on week2
#Nan represents absent ID
#customer 97 ordered food 4 on week1 and nothing on week 2
```

```python
len(week1.merge(week2,how="outer",on="Customer ID",suffixes=["- Week1", " - Week2"]))
# 454 rows exist compared to 500
#46 rows were those that matched Food IDs together in inner join
```

```python
week1.merge(week2,how="outer",on="Customer ID",suffixes=["- Week1", " - Week2"],
            indicator=True)
#indicator parameter gives visual value, shows where data is coming from 
```

```python
merged=week1.merge(week2,how="outer",on="Customer ID",suffixes=["- Week1", " - Week2"],
            indicator=True)
#assigning it to a variable
```

```python
merged["_merge"].value_counts()
```

```python
merged
#if we want only those values that are in either, we cut out rows with 'both' in merge column
```

```python
merged["_merge"].isin(["left_only","right only"])
```

```python
mask=merged["_merge"].isin(["left_only","right only"])
merged[mask]
#we have transitioned to data which is in either dataframe but not both
```

### Left Joins

129. Left Joins

```python
week1=pd.read_csv("Restaurant - Week 1 Sales.csv")
week2=pd.read_csv("Restaurant - Week 2 Sales.csv")
customers=pd.read_csv("Restaurant - Customers.csv")
foods=pd.read_csv("Restaurant - Foods.csv")
```

```python
#we set one dataframe as our main focus
#we look for values in A dataframe, values within dataframe A
#values exclusively in dataframe B will be ignored, if values exist in A as well, then it will be counted
#right join is the opposite of left join, but basically the same operation
```

```python
#it operates similar to vlookup operation
week1.head(3)
```

```python
foods.head(3)
```

```python
#we want food ID from week1 to reference in foods table
week1.merge(foods,how="left",on="Food ID")
#left dataframe is week1 dataframe
#it drags food ID in foods table and bring it in new dataset
#if food ID existed in week1 but not in foods, it would provide null value
#in our scenario, this isn't the case
```

```python
week1.merge(foods,how="left",on="Food ID",sort=True)
#sort parameter sorts our matching column which is Food ID
#brand new dataframe 
```

```python
week1=week1.merge(foods,how="left",on="Food ID",sort=True)
#same column name of Food ID in this scenario
```

```python
week1.head()
```

### The left_on and right_on Parameters

130. The left_on and right_on Parameters

```python
week1=pd.read_csv("Restaurant - Week 1 Sales.csv")
week2=pd.read_csv("Restaurant - Week 2 Sales.csv")
customers=pd.read_csv("Restaurant - Customers.csv")
foods=pd.read_csv("Restaurant - Foods.csv")
```

```python
#what happens when we dont have matching column names
#we use parameters called left_on and right_on
```

```python
week2.head()
```

```python
customers.head(3)
```

```python
#here in week2 table we have customer ID column
#in customers table, we have column named ID which is the same
```

```python
#we have to clarify which columns we want to merge on in each one
week2.merge(customers,how="left",left_on="Customer ID",right_on="ID")
#with on parameter, we got one column but here both columns are pulled
```

```python
week2.merge(customers,how="left",left_on="Customer ID",right_on="ID").drop("ID",axis="columns")
#ID column has been dropped from our dataframe
```

```python
week2.merge(customers,how="left",left_on="Customer ID",right_on="ID",sort=True).drop("ID",axis="columns")
#sorting done based on Customer ID column
```

```python
week2=week2.merge(customers,how="left",left_on="Customer ID",right_on="ID",sort=True).drop("ID",axis="columns")
week2.head()
```

### Merging by Indexes with the left_index and right_index Parameters

131. Merging by Indexes with the left_index and right_index Parameters

```python
week1=pd.read_csv("Restaurant - Week 1 Sales.csv")
week2=pd.read_csv("Restaurant - Week 2 Sales.csv")
customers=pd.read_csv("Restaurant - Customers.csv")
foods=pd.read_csv("Restaurant - Foods.csv")
```

```python
#sometime you want to merge based on index values instead of column values
#it can be a combination, merging based on taking left and right index and merge them together
#or take left dataframe index and right dataframe column - a combination
```

```python
customers.head()
```

```python
customers=pd.read_csv("Restaurant - Customers.csv",index_col="ID")
customers
```

```python
foods=pd.read_csv("Restaurant - Foods.csv",index_col="Food ID")
foods.head(3)
```

```python
customers.head(2)
```

```python
week1.head()
```

```python
week1.merge(customers,how="left",left_on="Customer ID",right_index=True)
#right_index True makes merging based on right daaframe index
#duplicate columns do not occur here
#ID column from right dataframe has been discarded 
```

```python
sales=week1.merge(customers,how="left",left_on="Customer ID",right_index=True)
sales
```

```python
#now we want to merge based on Food ID column in sales dataframe with foods Dataframe
sales.merge(foods,how="left",left_on="Food ID",right_index=True)

#similar process below using right dataframe column and left dataframe index
#sales.merge(foods,how="left",right_on="Food ID",left_index=True)
```

```python
sales=sales.merge(foods,how="left",left_on="Food ID",right_index=True)
sales.head(3)
```

```python
#lets merge by both indexes in dataframe
week1.head(3)
```

```python
week2.head(3)
```

```python
#here we want to merge on pandas numeric index
week1.merge(week2,how="left",left_index=True,right_index=True)
```

```python
week1.merge(week2,how="left",left_index=True,right_index=True,suffixes=[" - Week 1","- Week 2"])
```
